```python
from random import shuffle, randrange
import random

def scenraio_generator():
    scen = ["collect", "check_in"]
    
    shuffle(scen)

    if randrange(0,3) == 1:
        scen.append("find")

#     shuffle(scen)

    if randrange(1, 10) > 5:
        scen.insert(0, "greeting")

    if "check_in" in scen:
        
        if randrange(1, 10) > 4:
            scen.insert(scen.index("check_in") + 1, "ask")
            
        if randrange(1, 10) > 4:
            scen.insert(scen.index("check_in") + 1, "overturn")

    return scen
```
